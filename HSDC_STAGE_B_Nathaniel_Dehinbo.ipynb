```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_absolute_error
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score


x = df.drop(columns=['date','lights','Appliances'])
y= df['Appliances']
```

```python
x
```

```python
y
```

```python
x_training, x_testing, y_training, y_testing = train_test_split(x, y, test_size=0.3, random_state=42)
```

```python
scaler = MinMaxScaler()
x_training_scaled = scaler.fit_transform(x_training)
x_testing_scaled = scaler.transform(x_testing)
```

```python
model = LinearRegression()
model.fit(x_training_scaled, y_training)
```

```python
y_pred_training = model.predict(x_training_scaled)
y_pred_testing = model.predict(x_testing_scaled)
```

```python
mae_training = mean_absolute_error(y_training, y_pred_training)
print("Mean Absolute Error on training set:", round(mae_training, 3))
```

```python
rmse_training = mean_squared_error(y_training, y_pred_training, squared=False)
print("RMSE on training set:", round(rmse_training, 3))
```

```python
mae_testing = mean_absolute_error(y_testing, y_pred_testing)
print("Mean Absolute Error on testing set:", round(mae_testing, 3))
```

```python
rmse_testing = mean_squared_error(y_testing, y_pred_testing, squared=False)
print("RMSE on testing set:", round(rmse_testing, 3))
```
